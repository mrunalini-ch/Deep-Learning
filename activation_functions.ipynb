***compare the outputs of activation functions on a dataset.(Experiment 4)***

**key features**

1. activation functions
2. normalization for better visualization of the acivation functions.
3. loading the titanic dataset
4. outputs the comparisions



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Titanic dataset
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic = pd.read_csv(url)

#selecting a feature(here, Fare)
titanic['Fare'] = titanic['Fare'].fillna(titanic['Fare'].mean())
fare = titanic['Fare'].values  # Extract the 'Fare' column

# normalization
fare_normalized = (fare - np.min(fare)) / (np.max(fare) - np.min(fare))

#activation_functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

#applying af's
sigmoid_output = sigmoid(fare_normalized)
relu_output = relu(fare_normalized)
tanh_output = tanh(fare_normalized)

#plotting
plt.figure(figsize=(10, 6))

# Sigmoid
plt.plot(fare_normalized, sigmoid_output, label="Sigmoid", color="blue", marker='o', linestyle='none')

# ReLU
plt.plot(fare_normalized, relu_output, label="ReLU", color="green", marker='x', linestyle='none')

# Tanh
plt.plot(fare_normalized, tanh_output, label="Tanh", color="red", marker='s', linestyle='none')

# Configure plot
plt.title("Comparison of Activation Functions on Titanic Dataset (Normalized Fare)")
plt.xlabel("Normalized Input (Fare)")
plt.ylabel("Output")
plt.legend()
plt.grid()

# Show plot
plt.show()
```
